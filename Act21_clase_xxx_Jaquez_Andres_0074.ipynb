```python
# Clase Programador: Definición de la estructura
class Programador:
    # El método __init__ define los atributos de la clase
    def __init__(self, nombre, edad, experiencia):
        self.nombre = nombre        # Atributo 1
        self.edad = edad            # Atributo 2
        self.experiencia = experiencia  # Atributo 3

    # Función 1: Acción de programar
    def programar(self):
        print(f"{self.nombre} está escribiendo código en Python...")

    # Función 2: Acción de corregir errores
    def reparar_bug(self):
        print(f"{self.nombre} está eliminando errores del sistema. ¡Casi listo!")

    # Función 3: Acción de descansar/recargar energía
    def tomar_cafe(self):
        print(f"{self.nombre} está tomando café para recuperar energía.")

# --- Creación de Objetos (Instancias) ---

# Programador 1: Datos de ejemplo
programador1 = Programador("Ana García", 25, "2 años")

# Programador 2: Datos de ejemplo
programador2 = Programador("Luis Pérez", 40, "15 años")

# --- Ejecución de las funciones y muestra de datos ---

print(f"--- Ficha del Programador 1 ---")
print(f"Nombre: {programador1.nombre}")
print(f"Experiencia: {programador1.experiencia}")
programador1.programar()
programador1.tomar_cafe()

print("\n" + "-"*30 + "\n")

print(f"--- Ficha del Programador 2 ---")
print(f"Nombre: {programador2.nombre}")
print(f"Edad: {programador2.edad} años")
programador2.reparar_bug()
programador2.tomar_cafe()
print("Programa terminado")
print("Programa creado por Jaquez Andres NL= 27")
```
